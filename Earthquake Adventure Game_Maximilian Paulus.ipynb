```python
""" 
    Text Adventure Game developed for DAT-5301 Data Science: Python.
    The course is part of the MS. in Business Analytics at HULT International
    Business School.
    
    Author: Maximilian Paulus
    Submission Date: October 17, 2019"""

""" 
DocString:
    a) Introduction:
    This game is as earthquake survival game guiding the player through an
    earthquake scenario in the Bay Area by testing his / her knowledge 
    around Earthquakes. By answering questions, the player proceeds to 
    different stages of the game with the goal to survive the earthquake.
    
    b) Known Errors or Bugs:
        Function random_question() since random number is drawn every iteration,
        user could get the same question twice, even if he answered correctly.
"""

import time
import random
import sys

def typing(message):
    for letter in message:
        print(letter, end='')
        time.sleep(0.025)
        
        
def game_start():

    
    # asking for name (optional: and occupation)
    player_name = input("Welcome to the Bay area! What is your Name?\n\n")
    
    # asking for choice of living in the bay area
    loc = 1
    while loc > 0:
        location = input(f"""
Hey {player_name}, where in the bay do you live or would you like to live?

    a) San Francisco
    b) East Bay
        
    """)
        if location.lower() == 'a' or 'san francisco' in location.lower() or "sf" in location.lower() or location == "1":
            loc_sf = "\nWOW! San Francisco, that is a great choice\n"
            typing(loc_sf)
            break
    
        elif location.lower() == 'b' or 'east bay' in location.lower() or location == "1":
            loc_east = "\nThe East Bay huh? There is nice places out there!\n"
            typing(loc_east)
            break 
        else:
            loc_error = "\nOh, something went wrong. Please try again\n"
            typing(loc_error)
    
    # describing the earthquake happening (interupt print with random breaks 
    # to simulate earthquake)
    # asking first question on where to hide in order to get game started
    
    input("<Press enter to continue>\n")
    
    message_earthquake = """

It is a beautiful sunny day and you are having a coffee around 
the corner from your new place. But then, all of a sudden, the earth
begins shaking....an earthquake happens...the ground is bouncing,
coffee mugs fall from tables, people are screaming. \n
    """
    typing(message_earthquake)

    input("<Press enter to continue>\n")
 
    ### Question No. 1, first reaction to the earthquake ###
    
# 'answer' is used as an iterator for the while loop    
    answer = 2
# The while loop repeats the quesion 2 times, in case the user input is wrong
    while answer > 0:
        question_1 = input(f"""
The ground is still shaking and you should make sure to be safe. What do you do?
(Please enter the letter of the right answer. You can try {answer} more time(s).)

    a) Stop, Drop, and Roll?
    b) Drop, Cover, and Hold On?
    
    """)
        if question_1.lower() == 'b' or question_1.lower() in 'hold on?':
            message_q1 = """
Correct answer, you survive the first shock. You are heading 
straight home to see what the earthquake did to your apartment.\n         
            """
            typing(message_q1)
            break
        elif question_1.lower() == 'a' or question_1.lower() in 'roll':
            message_q1_wrong = "Wrong answer! Try again, this is an important lesson to learn!\n"
            answer -= 1
            
            if answer == 0:
                fail()
            else:
                typing(message_q1_wrong)
            
        else:
            message_q1_error = "Oops something went wrong! Please make sure to input the letter of the correct answer!\n"
            answer -= 1
            
            if answer == 0:
                fail()
            else:
                typing(message_q1_error)
                
                  
    input("<Press enter to continue>\n")
    
    ### Question No. 2, after arrival on the apartment ###
    
# 'answer' is used as an iterator for the while loop
    answer = 1
# the while loop with the condition 'answer > 0' in this case does not add much
# value but is used in case the number of possible wrong answers ('answer') is increased
    while answer > 0:
        question_2 = input("""

You arrive at your apartment and it seems chaotic but safe to enter. 
But how do you survive in there for a couple days? 
Do you have enough food and water?
How much water should you plan for in case of an emergency like that?
If you answer wrong, you must go outside to get supplies. 

    a) 10 liter per person per day?
    b) 1 gallon per person per day?
    c) 1 liter per person per day?
    d) 10 gallon per person per day?

    """)
        if question_2.lower() == 'b':
            message_home = """
That is correct, you are well prepared. You can stay at home for a while.
make sure you shut down gas valves and check pipes for leakage! \n         
            """
            typing(message_home)
            inside()
            break
        else:
            message_home_wrong = "\nWrong answer! Try again, this is an important lesson to learn!\n"
            # typing(message_home_wrong) # not required in this version since the loop only allows for 1 chance
            answer -= 1
            
        message_go_outside = """
You did not seem prepared for the earthquake.
You need to go out and try to get food and water\n"""              
        typing(message_go_outside)
        correct_answer = outside()
        
        
    input("<Press enter to continue>\n")
    
    
# Defining the funciton 'outside'. This stage is reached when Question No. 2 is answered wrong
def outside():

    # The game continues in a scenario outside of the players apartment.
    message_outside = """
You are outside now. This is dangerous but you had no choice.
Without enough water and food you cannot survive inside until the crisis is over.
You should be careful out here. Who knows what happens next.
    
    """
    typing(message_outside)
    
    
    
    # A list with possible scenarios is created
    scenario_outside_lst = ['Law enforcement is at capacity and cannot protect the city anymore! People break into stores and rob other people. Answer the following question to stay safe.'
                            ,'The Earthquake caused a Tsunami, you should seek higher grounds! Answer the following question to stay safe.'
                            , 'The ground is shaking again, an aftershock! Answer the following question to stay safe.'
                            , 'A damaged power line caused a fire!  Answer the following question to escape the fire.']
    
    # Generating random number that is used to select a scenario from the list
    scenario_outside = random.randint(0, (len(scenario_outside_lst)-1))
    
    input("<press enter to continue>\n")
    
    # The randomly chosen scenario is printed
    typing(scenario_outside_lst[scenario_outside])
    
    # since 'outside' is an unsafer environment, 2 questions need to be answered
    # A random question is selected that needs to be answered in order to proceed
    correct = 1
    for correct in range(1,3):
        correct_answer = random_question()
        if correct_answer == True:
            correct += 1
            if correct == 3:
                win()
    
def inside():
    
    # The game continues in a scenario inside of the players apartment.
    message_inside = """
Gladly you are still inside. This place is safe and you
almost start to relax a little bit and shake off the panic. But don't get
too comfortable. No one knows what happens next!
    
    """
    typing(message_inside)
    scenario_inside_lst = ['The Earthquake caused a Tsunami, you should seek higher grounds! Answer the following question to stay safe.'
                            ,'All of a sudden a power outtage occurs, do you have a flashlight and batteries at home?  Answer the following question to find the flashlight.'
                            ,' An aftershock strucks, your building collapses. You need to evacuate immediately!  Answer the following question to get out of the building. '
                            ,'A damaged power line caused a fire!  Answer the following question to escape the fire.']
    
    # Generating random number that is used to select a scenario from the list
    scenario_inside = random.randint(0, (len(scenario_inside_lst)-1))
    
    input("<press enter to continue>\n")
    
    # The randomly chosen scenario is printed
    typing(scenario_inside_lst[scenario_inside])
    
    # A random question is selected that needs to be answered in order to proceed
    correct_answer = random_question()
    
    # If the value 'True' is returned from the function random_question the game is won!
    if correct_answer == True:
        win()
    
def random_question():
    answer = 3
    correct_answer = False # 'correct_answer' will be used as return value of the random question function

    question_lst = [['A 4.0 Earthquake releases how much more energy than a 2.0 earthquake?', '2 times', '20 times', '100 times', 'c', '100 times'],
                    ['In San Francisco, continuous cables are used in new concrete construction. What is the main purpose of these cables?', 'to prevent the buildings from shaking during an earthquake', 'to anchor a building into the rock below','to bring the building back to its original position after an earthquake','c','to anchor a building into the rock below','to bring the building back to its original position after an earthquake'],
                    ['More people are injured or killed by falling furniture than building damage','True','False','','a','True'],
                    ['How often should you replace perishable items in your emergency kit such as water, food, meds, and batteries?','Every 4 months','Every 6 months','Every year','c','every year'],
                    ['How long can you keep frozen foods in the freezer if the door is closed?','Hours','One Night','A few days','c','a few days'],
                    ['From where shouldn\'t you retreive water if yours is tainted?','Water heaters','Radiators','Toilet Tank','b','radiators'],
                    ['How long after a major earthquake can aftershocks continue to happen?','Weeks','Days','Hours','a','weeks'],
                    ['When is the San Francisco Tsunami warning system tested?','Every first sunday of the month','Every tuesday','every other wednesday','b','every tuesday'],
                    ['If you see a large fire you should','leave immediately','call fire department and leave immediately','try to put it out','b','call fire department and leave immediately'],
                    ['How many earthquakes happen in southern Calfornia every year?','1000','2500','10000','c','10000']] 


    
    while answer > 0:
    
        # Generating random number that is used to select a question from the list
        question_select = random.randint(0, (len(question_lst)-1))
    
        # Asking for user input and saving it in 'choice'. The question is sliced
        # from the list of questions.
        choice = input(f"""
{question_lst[question_select][0]}:
You have {answer} chance(s) left.
    a) {question_lst[question_select][1]}
    b) {question_lst[question_select][2]}
    c) {question_lst[question_select][3]}

    """)
    
        # The user input is checked and an output is printed. In case the user answers
        # wrong, he gets a new question. Due to the random question selection he could end up with the same question (BUG!!!!!)
        if choice == question_lst[question_select][4]: 
            print(random.choice(['\nGreat, you are getting a hang of it!','\nAwesome, that was correct!','\nWell played, you answered correctly!']))
            correct_answer = True
            break
        else:
            message_incorrect = "\nSorry that was incorrect, please try again"
            typing(message_incorrect)
            answer -= 1
        # second if statement to see whether loop was interrupted due to correct answer or wrong answer
        if answer == 0:
            fail()
       
    return correct_answer

def win():
    again_iterator = 1
    message_win = """
Congratulations, you survived the crisis....well at least the one 
in this game. Since you made it to this point and answered the questions,
you are now prepared for a real crisis. But let's hope you will never need it!"""
    typing(message_win)
    while again_iterator > 0:
        
        again = input("""
Would you like to play again?
    a) yes
    b) no
    
    """)
        
        if again.lower() == "a" or again.lower() in "yes":
            game_start()
            break
        elif again.lower() == "b" or again.lower() in "no":
            sys.exit()
        else:
            print("\nSomething went wront. Please input what you want to do\n")
    
    
def fail():
    again_iterator = 1
    message_game_over = """
            
Your knowledge around earthquakes was not sufficient to survive the crisis. 
Sorry the game is over. You should read more about earthquakes!!!

              """
    typing(message_game_over)
    while again_iterator > 0:
        
        again = input("""
Would you like to play again?
    a) yes
    b) no
    
    """)
        
        if again.lower() == "a" or again.lower() in "yes":
            game_start()
            break
        elif again.lower() == "b" or again.lower() in "no":
            sys.exit()
            break
        else:
            print("\nSomething went wront. Please input what you want to do\n")
    
    
################################
          #Game Start#
################################

game_start()
```
